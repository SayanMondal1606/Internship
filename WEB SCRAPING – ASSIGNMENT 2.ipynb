```python
# [Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You
#have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
#jobs data.
#This task will be done in following steps:
#1. First get the webpage https://www.shine.com/
#2. Enter “Data Analyst” in “Job title, Skills” field and enter “Bangalore” in “enter the location” field.
#3. Then click the searchbutton.
#4. Then scrape the data for the first 10 jobs results you get.
#5. Finally create a dataframe of the scraped data.]
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from time import sleep

# Opening the Homepage
driver = webdriver.Chrome()
driver.get('https://www.shine.com/')

# Finding and entering job title
job_title = driver.find_element_by_id('search-keyword')
job_title.send_keys('Data Analyst')

# Finding and entering location
location = driver.find_element_by_id('locAutoSugg')
location.send_keys('Bangalore')

# Clicking on search button
search_btn = driver.find_element_by_xpath('//button[@class="btn search-btn"]')
search_btn.click()
sleep(5)

# Scraping details of the 10 jobs
jobs = driver.find_elements_by_xpath('//div[@class="jobTupleHeader"]')
for i in range(10):
    job_title = jobs[i].find_element_by_tag_name('a').text
    company = jobs[i].find_element_by_xpath('.//span[2]').text
    location = jobs[i].find_element_by_xpath('.//span[3]').text
    experience = jobs[i].find_element_by_xpath('.//span[4]').text
    salary = jobs[i].find_element_by_xpath('.//span[5]').text
    print(f'Job Title: {job_title}\nCompany: {company}\nLocation: {location}\nExperience: {experience}\nSalary: {salary}\n')

# Closing the page
driver.quit()
```

```python
#:Write a python program to scrape data for “Data Scientist” Job position in“Bangalore” location. You
#have to scrape the job-title, job-location, company_name. You have to scrape first 10 jobs data.
#This task will be done in following steps:
#1. First get the webpage https://www.shine.com/
#2. Enter “Data Scientist” in “Job title, Skills” field and enter “Bangalore” in “enter thelocation” field.
#3. Then click the search button.
#4. Then scrape the data for the first 10 jobs results you get.
#5. Finally create a dataframe of the scraped data.
```

```python
# Import libraries
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

# Set up the driver
driver = webdriver.Chrome()

# Visit website
driver.get("https://www.shine.com/")

# Enter job title
search_job = driver.find_element_by_name("qp")
search_job.send_keys("Data Scientist")

# Enter location
search_location = driver.find_element_by_xpath('//*[@id="ql"]/div/div/div/div[2]/div[1]/div/input')
search_location.send_keys("Bangalore")

# Click search button
search_button = driver.find_element_by_xpath('//*[@id="search_form"]/div[1]/div[3]/button')
search_button.click()

# Wait for the page to load
time.sleep(5)

# Scrape data for first 10 jobs
jobs = driver.find_elements_by_xpath('//li[@class="search__result search__result--v2 search__result--v2__list js-shine-list"]')[:10]

job_data = []

for job in jobs:
    # Get job title
    try:
        job_title = job.find_element_by_xpath('.//h3').text
    except:
        job_title = 'N/A'
        
    # Get job location
    try:
        job_location = job.find_element_by_xpath('.//span').text
    except:
        job_location = 'N/A'
        
    # Get company name
    try:
        company_name = job.find_element_by_xpath('.//span').text
    except:
        company_name = 'N/A'
        
    # Append job data to list
    job_data.append({'Job Title': job_title,
                     'Job Location': job_location,
                     'Company Name': company_name})

# Create dataframe from job data
df = pd.DataFrame(job_data)

# Print dataframe
print(df)

# Close the driver
driver.close()
```

```python
:# In this question you have to scrape data using the filters available on the webpage 
# You have to use the location and salary filter.
#You have to scrape data for “Data Scientist” designation for first 10 job results.
#You have to scrape the job-title, job-location, company name, experience required.
#The location filter to be used is “Delhi/NCR”. The salary filter to be used is “3-6” lakhs
#The task will be done as shown in the below steps:
#1. first get the web page https://www.shine.com/
#2. Enter “Data Scientist” in “Skill, Designations, and Companies” field.
#3. Then click the search button.
#4. Then apply the location filter and salary filter by checking the respective boxes
#5. Then scrape the data for the first 10 jobs results you get.
#6. Finally create a dataframe of the scrapeddata
```

```python
# Import libraries
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Set up the driver
driver = webdriver.Chrome()

# Visit website
driver.get("https://www.shine.com/")

# Enter job title and click search button
search_job = driver.find_element_by_name("qp")
search_job.send_keys("Data Scientist")
search_button = driver.find_element_by_xpath('//*[@id="search_form"]/div[1]/div[3]/button')
search_button.click()

# Wait for the page to load
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_element_located((By.XPATH, '//div[contains(@class,"filter__group no-selection")]')))

# Click location filter and select Delhi/NCR
location_filter = driver.find_element_by_xpath('//span[text()="Location"]/parent::li')
location_filter.click()
delhi_filter = driver.find_element_by_xpath('//span[text()="Delhi / NCR"]')
delhi_filter.click()

# Click salary filter and select 3-6 lakhs
salary_filter = driver.find_element_by_xpath('//span[text()="Salary"]/parent::li')
salary_filter.click()
salary_range = driver.find_element_by_xpath('//label[text()="3-6 Lakhs"]')
salary_range.click()

# Click apply button
apply_button = driver.find_element_by_xpath('//button[text()="Apply"]')
apply_button.click()

# Wait for the page to load
time.sleep(5)

# Scrape data for first 10 jobs
jobs = driver.find_elements_by_xpath('//li[@class="search__result search__result--v2 search__result--v2__list js-shine-list"]')[:10]

job_data = []

for job in jobs:
    # Get job title
    try:
        job_title = job.find_element_by_xpath('.//h3').text
    except:
        job_title = 'N/A'
        
    # Get job location
    try:
        job_location = job.find_element_by_xpath('.//span').text
    except:
        job_location = 'N/A'
        
    # Get company name
    try:
        company_name = job.find_element_by_xpath('.//span').text
    except:
        company_name = 'N/A'
    
    # Get experience required
    try:
        exp = job.find_element_by_xpath('.//li[1]').text
    except:
        exp = 'N/A'
    
    # Append job data to list
    job_data.append({'Job Title': job_title,
                     'Job Location': job_location,
                     'Company Name': company_name,
                     'Experience Required': exp})

# Create dataframe from job data
df = pd.DataFrame(job_data)

# Print dataframe
print(df)

# Close the driver
driver.close()
```

```python
# Import libraries
import pandas as pd
import time
from selenium import webdriver

# Set up the driver
driver = webdriver.Chrome()

# Visit website
driver.get("https://www.flipkart.com/")

# Close login popup if it appears
try:
    close_button = driver.find_element_by_xpath('//button[@class="_2KpZ6l _2doB4z"]')
    close_button.click()
except:
    pass

# Search for sunglasses and click search button
search_item = driver.find_element_by_name("q")
search_item.send_keys("sunglasses")
search_button = driver.find_element_by_xpath('//button[@class="L0Z3Pu"]')
search_button.click()

# Wait for the page to load
time.sleep(5)

# Scrape data for first 100 sunglasses
sunglasses_data = []
while len(sunglasses_data) < 100:
    sunglasses = driver.find_elements_by_xpath('//a[@class="_1fQZEK"]')
    for sunglass in sunglasses:
        # Click on sunglass listing and switch to new window
        sunglass.click()
        driver.switch_to.window(driver.window_handles[1])
        
        # Get brand
        try:
            brand = driver.find_element_by_xpath('//span[text()="Brand"]/following-sibling::span').text
        except:
            brand = 'N/A'
        
        # Get product description
        try:
            prod_desc = driver.find_element_by_xpath('//h1').text
        except:
            prod_desc = 'N/A'
        
        # Get price
        try:
            price = driver.find_element_by_xpath('//div[@class="_30jeq3 _16Jk6d"]').text
        except:
            price = 'N/A'
        
        # Append sunglass data to list
        sunglasses_data.append({'Brand': brand,
                                'Product Description': prod_desc,
                                'Price': price})
        
        # Close current window and switch back to original window
        driver.close()
        driver.switch_to.window(driver.window_handles[0])
        
        # Wait for a few seconds before clicking the next sunglass listing
        time.sleep(2)
    
    # Click next page button
    try:
        next_button = driver.find_element_by_xpath('//a[@class="_1LKTO3"][text()="Next"]')
        next_button.click()
    except:
        break
    
    # Wait for the page to load
    time.sleep(5)

# Create dataframe from sunglass data
df = pd.DataFrame(sunglasses_data)

# Print dataframe
print(df)

# Close the driver
driver.close()
```

```python
#Scrape 100 reviews data from flipkart.com for iphone11 phone. You have to go the link:
#https://www.flipkart.com/apple-iphone-11-black-64-gb/productreviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&market
#place=FLIPKART
```

```python

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# set up the webdriver
driver_path = 'path/to/chromedriver'
service = Service(driver_path)
options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--ignore-ssl-errors')
driver = webdriver.Chrome(service=service, options=options)

# get the webpage
url = 'https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART'
driver.get(url)

# close the login popup, if it appears
try:
    close_popup = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//button[@class="_2KpZ6l _2doB4z"]')))
    close_popup.click()
except:
    pass

# scrape data from the reviews
reviews = driver.find_elements_by_xpath('//div[@class="_1AtVbE col-12-12"]')
count = 1
for review in reviews:
    if count > 100:  # limit to first 100 reviews
        break
        
    # scrape the rating
    try:    
        rating = review.find_element_by_xpath('.//div[contains(@class,"_3LWZlK")]/text()')
    except:
        rating = 'N/A'

    # scrape the review summary
    try:
        summary = review.find_element_by_xpath('.//p[contains(@class,"_2-N8zT")]/text()')
    except:
        summary = 'N/A'

    # scrape the full review
    try:
        readmore = review.find_element_by_xpath('.//div[@class="_3AT_6t _1tsVs4"][contains(text(),"READ MORE")]/parent::node()')
        readmore.click()
        fullreview = review.find_element_by_xpath('.//div[@class="_1PBCrt"]/text()')
    except:
        try:
            fullreview = review.find_element_by_xpath('.//div[@class="_1PBCrt"]/text()')
        except:
            fullreview = 'N/A'

    # print the scraped data
    print(f'{count}. Rating: {rating}  Summary: {summary}  Full Review: {fullreview}\n')
    count += 1

# close the webdriver
driver.quit()
```

```python
#Q6: Scrape data forfirst 100 sneakers you find whenyou visit flipkart.com and search for “sneakers” inthe
#search field.
#You have to scrape 3 attributes of each sneaker:
#1. Brand
#2. ProductDescription
#3. Price
```

```python
# Import required libraries
import time
from selenium import webdriver

# Set up the driver
driver = webdriver.Chrome()

# Visit flipkart.com
driver.get('https://www.flipkart.com/')

# Close login popup if displayed
try:
    driver.find_element_by_xpath("//button[text()='✕']").click()
except:
    pass

# Search for sneakers and click search button
search_box = driver.find_element_by_name('q')
search_box.send_keys('sneakers')
search_button = driver.find_element_by_class_name('L0Z3Pu')
search_button.click()

# Wait for the page to load
time.sleep(5)

# Scrape data for first 100 sneakers
for page in range(1, 6):
    # Wait for the page to load
    time.sleep(2)
    
    # Get list of sneakers on the page
    sneakers = driver.find_elements_by_xpath("//div[@class='_2kHMtA']")
    
    # Loop through each sneaker and scrape data
    for sneaker in sneakers:
        # Get brand
        brand = sneaker.find_element_by_xpath(".//div[@class='_2B_pmu']").text
        
        # Get product description
        product_desc = sneaker.find_element_by_xpath(".//a[1]").get_attribute('title')
        
        # Get price
        price = sneaker.find_element_by_xpath(".//div[@class='_30jeq3 _1_WHN1']").text
        
        # Print data
        print('Brand:', brand)
        print('Product Description:', product_desc)
        print('Price:', price)
        print('')
        
    # Go to the next page
    try:
        next_button = driver.find_element_by_xpath("//a[text()='Next']").click()
    except:
        break

# Close the driver
driver.close()
```
