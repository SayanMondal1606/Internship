# Model testing

```python
X=df.iloc[:, 1:43].values

Y=df.iloc[:, 42].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.20, random_state=1, shuffle= True)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
```

```python
# Define and train various machine learning models
models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(alpha=1),
    "Lasso Regression": Lasso(alpha=1),
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(n_estimators=100, random_state=42),
    "Support Vector Regression": SVR(kernel="rbf")
}

# Train models and evaluate performance
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    print(f"{name}: Mean Squared Error = {mse:.4f}")
```

```python

# Residual analysis
for model in models.values():
    y_pred = model.predict(X_test)
    residuals = y_test - y_pred

    # Visualize residuals
    plt.scatter(y_pred, residuals)
    plt.xlabel("Predicted CGPA")
    plt.ylabel("Residuals")
    plt.title("Residual Plot for {}".format(model.__class__.__name__))
    plt.show()
from sklearn.metrics import r2_score

# Prediction evaluation
for name, model in models.items():
    y_pred = model.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    print(f"{name}: R-squared = {r2:.4f}")

    # Compare actual and predicted CGPA values
    plt.figure(figsize=(8, 6))
    plt.scatter(y_test, y_pred)
    plt.xlabel("Actual CGPA")
    plt.ylabel("Predicted CGPA")
    plt.title("Actual vs. Predicted CGPA for {}".format(name))
    plt.plot([0, 10], [0, 10], color="red", linestyle="--")  # Identity line
    plt.show()
```
